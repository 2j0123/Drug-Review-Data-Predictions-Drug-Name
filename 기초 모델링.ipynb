```python
import pickle
import pandas as pd
```

```python
#  word_encoding  
with open('word_encoding.pickle', 'rb') as file:
    loaded_word_encoding = pickle.load(file)

#   
print(loaded_word_encoding)
```

```python
df = pd.read_csv("/content/drug_eda.tsv", sep = '\t')
```

```python
#   
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

import tensorflow_datasets as tfds
import tensorflow as tf
```

```python
from sklearn.preprocessing import LabelEncoder
LEx = LabelEncoder()
LEy = LabelEncoder()
```

```python
X=df['condition']
y=df['drugName']
```

```python
LEx.fit(X)
LEy.fit(y)
```

```python
x_encoded = LEx.transform(X)
y_encoded = LEy.transform(y)
a=pd.DataFrame(x_encoded,columns=['x_encoded'])
b=pd.DataFrame(y_encoded, columns=['y_encoded'])

df=pd.concat([df,a,b],axis=1)
```

```python
X.iloc[0]
```

```python
def encoding(words):
    numbers = [loaded_word_encoding[word] for word in words if word in loaded_word_encoding]
    return numbers
df['encoded_review'] =  df['tokenized_review'].apply(encoding)
```

```python
df['x_encoded'] = df['x_encoded'].apply(lambda x: [x])
```

```python
for i in range(len(df['encoded_review'])):
    df['x_encoded'][i].extend(df['encoded_review'][i])
```

```python
print(df.head(5))
```

```python
X_pad = pad_sequences(df['x_encoded'],maxlen=200) #   : 159  1  3712
```

```python
type(X_pad)
```

```python
print(X_pad[0])
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
```

```python
X_pad = X_pad.astype(np.float64)
```

```python
y_onehot = to_categorical(df['y_encoded'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pad, y_onehot, test_size=0.2, random_state=42)
```

```python
model_fnn = Sequential()
model_fnn.add(Embedding(input_dim=len(loaded_word_encoding)+1, output_dim=256, input_length=200))
model_fnn.add(Flatten())
model_fnn.add(Dense(2048, activation='relu'))
model_fnn.add(Dense(1024, activation='relu'))
model_fnn.add(Dense(3175, activation='softmax'))
```

```python
model_fnn.summary()
```

```python
model_fnn.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Nadam(lr=0.001), metrics=['accuracy'])
```

```python
model_fnn.fit(X_train, y_train, epochs=10, batch_size=128, validation_split=0.2)
```

```python
score = model_fnn.evaluate(X_test, y_test)
print("\nTest score:", score[0])
print('Test accuracy:', score[1])
```

```python
input_data=pad_sequences([df['x_encoded'].iloc[0]],maxlen=200)
output = model_fnn.predict(input_data)
print(output)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
model_rnn = Sequential()
model_rnn.add(Embedding(input_dim=len(loaded_word_encoding)+1, output_dim=256, input_length=200))

#
model_rnn.add(SimpleRNN(256, return_sequences=False)) # .

#dense 
model_rnn.add(Dense(256, activation='relu'))
model_rnn.add(Dense(128, activation='relu'))
model_rnn.add(Dense(3175, activation='softmax'))
```

```python
model_rnn.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Nadam(lr=0.0005), metrics=['accuracy'])
```

```python
model_rnn.fit(X_train, y_train, epochs=10, batch_size=128, validation_split=0.2)
```

```python
score = model_rnn.evaluate(X_test, y_test)
print("\nTest score:", score[0])
print('Test accuracy:', score[1])
```
